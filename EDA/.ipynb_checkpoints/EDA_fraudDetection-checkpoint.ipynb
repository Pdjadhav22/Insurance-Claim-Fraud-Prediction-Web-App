```python
# policy_state vs target variable
policy_stateCT =  pd.crosstab(cat_df['policy_state'],cat_df['fraud_reported'],margins=True)
policy_stateCT
```

```python
from scipy.stats import chi2_contingency
chiResults = chi2_contingency(policy_stateCT)
chiResults
```

```python
# function for calculating chi2 stat & p-Value for each cat features with respect to Target
def chi2Test(X,y):    
    cT = pd.crosstab(X,y)
    chiRes = chi2_contingency(cT)
    chiStat = chiRes[0]
    pValue = chiRes[1]
    return chiStat, pValue
```

```python
# Getting chi2 test results for each cat col
chi2Results = pd.DataFrame(columns=["col_name","chi2_stat",'pValue'])
catSelected = []
for col in cat_df.iloc[:,:-1].columns:
    x = cat_df[col]
    y = cat_df['fraud_reported']
    stat, pValue = chi2Test(x,y)
#     print(f"{col}-->{stat},{pValue}")
    res_df = pd.DataFrame([[col,stat,pValue]],columns=["col_name","chi2_stat",'pValue']) 
    chi2Results = pd.concat([chi2Results,res_df], ignore_index=True)
    if pValue < 0.05:
        catSelected.append(col)
chi2Results
```

```python
# Final selected Cat Variables
cat_df_selected = data[catSelected]
cat_df_selected
```
